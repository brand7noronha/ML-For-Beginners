```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

# loading and checking the shape of the dataset
X, y = datasets.load_diabetes(return_X_y=True)
X = X[:,[2]] # selecting only one feature i.e., the 3rd column of the dataset
X = X.reshape((-1,1)) # reshaping the data to be of 2D array for better plotting
print(X.shape)
# print(X[0])
```

```python
# splitting both the data (X) and target (y) into test and training sets
# or in general, splitting our data into training and testing data
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# we then create a linear regression model and train it with our data
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# predicting using the test data
y_pred = model.predict(X_test)
```

```python
# plotting the data using matplotlib
# creating a scatter plot
plt.scatter(X_test, y_test, color='black')

# plotting the predictions
plt.plot(X_test, y_pred, color='blue', linewidth=3)

# add labels and title
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')

# draw the plot
plt.show()
```
